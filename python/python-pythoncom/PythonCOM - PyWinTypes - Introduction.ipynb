```python
import pywintypes
import time
import datetime
```

```python

# create a time object
myTime = time.time()

# easy way to create a datetime object from 1970 to 1938
new_datetime = datetime.datetime.fromtimestamp(int (myTime))

# convert it to a Time object in PyWin
myTime = pywintypes.Time(myTime)

# what is the type
display(type(myTime))
display(myTime)

# display the parts of the pyWin Datetime
display(myTime.year)
display(myTime.month)
display(myTime.day)
display(myTime.hour)
display(myTime.minute)
display(myTime.second)

# convert the pyWinType to a normal datetime object
new_datetime = datetime.datetime (
  year=myTime.year,
  month=myTime.month,
  day=myTime.day,
  hour=myTime.hour,
  minute=myTime.minute,
  second=myTime.second
)
print(new_datetime)
```
